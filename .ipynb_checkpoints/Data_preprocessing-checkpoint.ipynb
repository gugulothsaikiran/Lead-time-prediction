```python
import pandas as pd #used for reading file
import numpy as np #used for making mathematical calculations on arrays
from IPython.display import display #to change design view of jupyter note book output
```

```python
data = pd.read_excel("student_project_Lead_time_prediction_data.xlsx") 

# path = input().replace("\\",'/')

# if path[-1]=='v':
#     data = pd.read_csv(path)
# else: 
#     data = pd.read_excel(path)

data.info()
```

```python
data.head()
```

```python
data_req = data.drop(['SH_CD','Unnamed: 11','CATALOG_DESC'],axis=1)
final_data = data_req.dropna()
final_data.info()
```

```python
final_data = final_data.sort_values('CATALOG_NO')
final_data.head(25)
```

```python
final_data['INDENT_DT'] = pd.to_datetime(final_data['INDENT_DT'],errors = 'coerce')
final_data['MATL_RECEIPT_DT'] = pd.to_datetime(final_data['MATL_RECEIPT_DT'],errors = 'coerce')
final_data = final_data.dropna()

final_data['CATALOG_NO'] = final_data['CATALOG_NO'].astype(np.int64)

final_data.to_csv('Processed_data.csv',index=False)
```
