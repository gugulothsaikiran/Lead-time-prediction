```python
import pandas as pd
import numpy as np
from IPython.display import display

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import LearningCurveDisplay
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse


import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
final_data = pd.read_csv('Processed_data.csv')
final_data.head()
```

```python
catalog_grouped_data = final_data.groupby(['CATALOG_NO']).count()
catalog_grouped_data = catalog_grouped_data[catalog_grouped_data['INDENT_DT']>9]
display(catalog_grouped_data.head(25))
unique_catalog_numbers = list(catalog_grouped_data.index)
```

```python
data = final_data[final_data['CATALOG_NO'].isin(unique_catalog_numbers)]
data['INDENT_DT'] = pd.to_datetime(data['INDENT_DT'])
data['MATL_RECEIPT_DT'] = pd.to_datetime(data['MATL_RECEIPT_DT'])
data.head()
```

```python
data['DAYS'] = (data['MATL_RECEIPT_DT'] - data['INDENT_DT']).dt.days
data.head()
```

```python
encoder = LabelEncoder()

data['AR_ITEM_IND_e'] = encoder.fit_transform(data['AR_ITEM_IND'])
data['SPC_ITEM_IND_e']  = encoder.transform(data['SPC_ITEM_IND']) 
data['QUALITY_IND_e']  = encoder.transform(data['QUALITY_IND']) 
data['FTP_IND_e']  = encoder.transform(data['FTP_IND']) 

catalog_encoder = LabelEncoder().fit(data['CATALOG_NO'])

data['CATALOG_NO_e'] = catalog_encoder.transform(data['CATALOG_NO'])
catalog_encoder.classes_
```

```python
dataset = data[['CATALOG_NO','AR_ITEM_IND_e','SPC_ITEM_IND_e','QUALITY_IND_e','FTP_IND_e','DAYS']] 

plt.figure(figsize=(8, 6))  # Set the size of the heatmap (optional)
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
X = data[['CATALOG_NO']]
Y = data['DAYS']
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20,random_state= 5)
```

```python
models = [LinearRegression(),DecisionTreeRegressor(),SGDRegressor(),RandomForestRegressor()]

for i in models:
    model = i.fit(x_train,y_train)
    train_er = i.predict(x_train)
    y_pred = i.predict(x_test)
    print(i)
    print("Training Eror")
    print("MSE:- ",mse(y_train,train_er))
    print("MAE:- ",mae(y_train,train_er))
    print("\nTesting Error")
    print("MSE:- ",mse(y_test,y_pred))
    print("MAE:- ",mae(y_test,y_pred),"\n\n")
    
    
```

```python
learning_curve = LearningCurveDisplay.from_estimator(DecisionTreeRegressor(),x_train,y_train,cv=5, n_jobs=4, train_sizes=np.linspace(0.01, 1.0, 50))
```

```python
grouped_df = data.groupby("CATALOG_NO")["DAYS"].mean()
plt.plot(grouped_df)
plt.xlabel("Product ID")
plt.ylabel("Average delivery time (days)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['INDENT_DT'], bins=30, alpha=0.7, label='Indent Dates')
plt.hist(data['MATL_RECEIPT_DT'], bins=30, alpha=0.7, label='Received Dates')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.legend()
plt.title('Distribution of Indent and Received Dates')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['INDENT_DT'] = pd.to_datetime(data['INDENT_DT'])
data['MATL_RECEIPT_DT'] = pd.to_datetime(data['MATL_RECEIPT_DT'])
orders_over_time =data.groupby('INDENT_DT').size()
orders_over_time.plot()
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Over Time')
plt.show()
```

```python
# Define the hyperparameters to tune
param_grid = {
    'max_depth': [None, 5, 10, 15, 20,40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Create the DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor()

# Initialize GridSearchCV with the DecisionTreeRegressor and the parameter grid
grid_search = GridSearchCV(dt_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')

# Separate features and target variables
X = data[['CATALOG_NO']]
y = data['DAYS']

# Perform the grid search to find the best hyperparameters
grid_search.fit(X, y)

# Get the best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print("Best Hyperparameters:", best_params)
print("Best Model:", best_model)

# Train the best model on the entire dataset
best_model.fit(X, y)

# Now you can use the best_model for predictions
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Assuming you have already trained the tuned DecisionTreeRegressor model and named it as "best_model"

# Separate features and target variables
X = data[['CATALOG_NO']]
y = data['DAYS']

# Make predictions on the training set and test set
y_train_pred = best_model.predict(X)
y_test_pred = best_model.predict(x_test)

# Evaluate the model on the training set
train_mse = mean_squared_error(y, y_train_pred)
train_mae = mean_absolute_error(y, y_train_pred)
train_r2 = r2_score(y, y_train_pred)

# Evaluate the model on the test set
test_mse = mean_squared_error(y_test, y_test_pred)
test_mae = mean_absolute_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)

# Print the evaluation metrics
print("Training Set Metrics:")
print("MSE:", train_mse)
print("MAE:", train_mae)
print("R2:", train_r2)

print("\nTest Set Metrics:")
print("MSE:", test_mse)
print("MAE:", test_mae)
print("R2:", test_r2)

# Visualize the actual vs. predicted values on the test set
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Days')
plt.ylabel('Predicted Days')
plt.title('Actual vs. Predicted Values')
plt.show()
```

```python
# Create instances of the models
svr_model = SVR()
gradient_boosting_model = GradientBoostingRegressor()
xgboost_model = xgb.XGBRegressor()

# Train the models on the training set
svr_model.fit(x_train, y_train)
gradient_boosting_model.fit(x_train, y_train)
xgboost_model.fit(x_train, y_train)

# Make predictions on the test set
y_svr_pred = svr_model.predict(x_test)
y_gradient_boosting_pred = gradient_boosting_model.predict(x_test)
y_xgboost_pred = xgboost_model.predict(x_test)

# Evaluate the models on the test set
svr_mse = mean_squared_error(y_test, y_svr_pred)
svr_mae = mean_absolute_error(y_test, y_svr_pred)
svr_r2 = r2_score(y_test, y_svr_pred)

gradient_boosting_mse = mean_squared_error(y_test, y_gradient_boosting_pred)
gradient_boosting_mae = mean_absolute_error(y_test, y_gradient_boosting_pred)
gradient_boosting_r2 = r2_score(y_test, y_gradient_boosting_pred)

xgboost_mse = mean_squared_error(y_test, y_xgboost_pred)
xgboost_mae = mean_absolute_error(y_test, y_xgboost_pred)
xgboost_r2 = r2_score(y_test, y_xgboost_pred)

# Print the evaluation metrics for each model
print("Support Vector Regression (SVR) Metrics:")
print("MSE:", svr_mse)
print("MAE:", svr_mae)
print("R2:", svr_r2)

print("\nGradient Boosting Regression Metrics:")
print("MSE:", gradient_boosting_mse)
print("MAE:", gradient_boosting_mae)
print("R2:", gradient_boosting_r2)

print("\nXGBoost Regression Metrics:")
print("MSE:", xgboost_mse)
print("MAE:", xgboost_mae)
print("R2:", xgboost_r2)
```
